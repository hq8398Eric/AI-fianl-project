```python
import pandas as pd
import numpy as np
import os
import random
from tqdm import tqdm
pd.options.mode.chained_assignment = None 

data_folder = "../data"
```

```python
data = pd.read_csv(f"{data_folder}/train/mirror_30.csv")
max_rallyid = max(data['rally_id'])
match_data = pd.read_csv(f"{data_folder}/train/augment_databyarea.csv")

p = 0.5

match_data
```

```python
if os.path.exists('./augumented_data.csv'):
    os.remove('./augumented_data.csv')
if os.path.exists('./concat_augumented_data.csv'):
    os.remove('./concat_augumented_data.csv')
    
new_data = pd.DataFrame(columns=data.columns)
cur = max_rallyid + 1

# for index, _ in tqdm(data.iterrows(), total = data.shape[0]):
#     if random.random() < p1:
#         new_data = new_data.append(data.loc[index])
    
for index, _ in tqdm(match_data.iterrows(), total = match_data.shape[0]):
    if random.random() < p:
        tmp = match_data.loc[index]
        tmp['rally_id'] = cur
        cur = cur + 1
        new_data = new_data.append(tmp)

new_data.to_csv('augumented_data.csv', index = False)

concat_data = pd.concat([data, new_data], ignore_index=True)
concat_data.to_csv('concat_augumented_data.csv', index = False)
```
