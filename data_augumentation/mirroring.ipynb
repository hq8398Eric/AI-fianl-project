```python
import pandas as pd
import numpy as np
import os
import random
pd.options.mode.chained_assignment = None 

data_folder = "../data"
```

```python
data = pd.read_csv(f"{data_folder}/train/train.csv")
max_rallyid = max(data['rally_id'])

p = 0.3

data
```

```python
if os.path.exists('./augumented_data.csv'):
    os.remove('./augumented_data.csv')
if os.path.exists('./concat_augumented_data.csv'):
    os.remove('./concat_augumented_data.csv')
    
new_data = pd.DataFrame(columns=data.columns)
d = False
cur = max_rallyid + 1
    
for index, _ in data.iterrows():
    if index == 0 or data.loc[index, 'rally_id'] != data.loc[index - 1, 'rally_id']:
        if random.random() < p:
            d = True
        else:
            d = False
    if d == True:
        tmp = data.loc[index].copy()
        
        if tmp['player_location_y'] < 500:
            tmp['player_location_y'] += 500
            tmp['opponent_location_y'] -= 500
        else:
            tmp['player_location_y'] -= 500
            tmp['opponent_location_y'] += 500
            
        tmp['landing_y'] *= -1
        tmp['rally_id'] = cur
        cur = cur + 1
        new_data = new_data.append(tmp, ignore_index = True) 

new_data.to_csv('augumented_data.csv', index = False)

concat_data = pd.concat([data, new_data], ignore_index=True)
concat_data.to_csv('concat_augumented_data.csv', index = False)
```
